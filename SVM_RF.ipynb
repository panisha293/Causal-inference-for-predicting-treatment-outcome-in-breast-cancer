```python
%%time
#libraries
# !pip install pca
import numpy as np
import pandas as pd
from pca import pca
import matplotlib.pyplot as plt
from scipy.stats import randint
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import RandomizedSearchCV, train_test_split,StratifiedKFold, LeaveOneOut, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
from sklearn.metrics import classification_report, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import StratifiedKFold, LeaveOneOut, cross_val_score

pd.set_option('display.max_columns', None)

#read the files
img_feat=pd.read_excel("Imaging_Features.xlsx")
clinical_other_feat=pd.read_excel("Clinical_and_Other_Features.xlsx")

#append the target column to image features dataframe
clinical_other_feat = clinical_other_feat.drop([0,1])
clinical_other_feat.reset_index(drop=True,inplace=True)
img_feat['Tumor_Response']= clinical_other_feat['Unnamed: 59']



#drop NA values of the target column
img_feat.dropna(subset=['Tumor_Response'],inplace=True)

#get only options 1 and 2 from target variable and selecting rows based on options 
options = [1,2] 
img_feat = img_feat[img_feat['Tumor_Response'].isin(options)] 



#filling null values with mean and the below variable has the dictioanry of column with with 
#number of null values - just for reference
null_column_dict = {} 
for column in img_feat.columns[img_feat.isnull().any()]:
    null_column_dict[column] = img_feat[column].isna().sum()
    img_feat[column] = img_feat[column].fillna((img_feat[column].mean()))

    
print(len(img_feat['Tumor_Response']))

X = img_feat.drop(['Patient ID','Tumor_Response'],axis=1)
y = img_feat['Tumor_Response']

#perform pca on entire dataset to see the top principal components
model = pca(verbose=False)
pca_PC = model.fit_transform(X)
# print(pca_PC['topfeat'])
model.plot()

y = y.astype('int')
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)

pca = PCA(n_components=3)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)

# hyperparameter tuning
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 200, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 100, num = 11)]
max_depth.append(None)
min_samples_split = [2, 5, 10, 15, 20]
min_samples_leaf = [4, 8, 16, 20, 40, 80, 100]
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

cv = LeaveOneOut()
rf = RandomForestClassifier(random_state = np.random.seed(1))
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, 
                            cv = cv, verbose=2, random_state=42, n_jobs = -1)

#fit in the model
rf_random.fit(X_train, y_train)
best_rf = rf_random.best_estimator_
print('Best hyperparameters:',  rf_random.best_params_)

#predictions
predictions = best_rf.predict(X_test)
errors = abs(predictions - y_test)
mape = 100 * np.mean(errors / y_test)
accuracy = 100 - mape

#get the accuracy and error
print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
print('Accuracy = {:0.2f}%.'.format(accuracy))
```

```python
cm = confusion_matrix(y_test, predictions, labels=best_rf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=best_rf.classes_)
disp.plot()
plt.show()
```
